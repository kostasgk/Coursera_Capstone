```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup

url = 'https://www.weatherdatasource.com/GR-Attiki'#retrieve data foa Attiki region from weatherdatasource 
website_text = requests.get(url).text
soup = BeautifulSoup(website_text,'xml')
table_contents = []
divs = soup.find_all("ul", {"class": "list-group w-100 align-items-stretch"})
                     
for ultag in divs:
     for litag in ultag.find_all('li'):
            cell = {}
            cell['City']=litag.text
            cell['State']='Attiki'
            table_contents.append(cell)
#print(table_contents)
df_CityList=pd.DataFrame(table_contents).sort_values(by='City')
```

Getting the coordinates for the city of Athens

```python
address = 'Athens, Attiki'

geolocator = Nominatim(user_agent="athens_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Athens, Attiki are {}, {}.'.format(latitude, longitude))
```

Printing all the areas

```python
list_lat = []   # create empty lists
list_long = []
#geolocator = Nominatim(user_agent="ny_explorer")
for index, row in df_CityList.iterrows():
    #City = row['City']
    #State = row['State']  
    address = str(row['City'])+', '+str(row['State'])
    #print(address)
    geolocator = Nominatim(user_agent="athens_explorer")
    location = geolocator.geocode(print(address))
    latitude = location.latitude
    longitude = location.longitude

    list_lat.append(latitude)
    list_long.append(longitude)

#create new columns from lists    
df_CityList['Latitude'] = list_lat   
df_CityList['Longtitude'] = list_long
df_CityList.head(10)
```

There are 92 cities or neighborhoods in the greater Attiki area

```python
import folium
map_Attiki = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_CityList['Latitude'], df_CityList['Longtitude'], df_CityList['State'], df_CityList['City']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Attiki)  
    
map_Attiki
```
